```python
import numpy as np
import matplotlib.pyplot as plt
from day19p1 import read_input
from itertools import cycle

scanners = read_input("test_input.txt")
len(scanners)
```

```python
colours = cycle(['red', 'green', 'blue', 'yellow', 'purple'])
fig = plt.figure()
ax = plt.axes(projection ='3d')
for scanner in scanners[0:2]:
    px = [p[0] for p in scanner._points]
    py = [p[1] for p in scanner._points]
    pz = [p[2] for p in scanner._points]

    ax.scatter3D(px, py, pz, c=next(colours))
    
plt.show()
```
